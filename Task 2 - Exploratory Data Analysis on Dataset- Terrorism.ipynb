## Data Cleaning 

```python
df.rename(columns={"iyear":'Year','imonth':'Month','iday':'Day','country_txt':'Country','extended':'Extended','region_txt':'Region',
          'provstate':'State','city':'City','attacktype1_txt':'Attack_Type','targtype1_txt':'Target_Type','weaptype1_txt':
          'Weapon','nkill':'Killed','nwound':'Wounded','gname':'Group'},inplace=True)
```

```python
df1=df[['Year','Month','Day','Country','Extended','Region','State','City','Attack_Type','Target_Type','Weapon','Killed',
        'Wounded','Group']]
df1
```

```python
df1.shape
```

```python
df1.describe()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1['Killed'].fillna(0,inplace=True)
df1['Wounded'].fillna(0,inplace=True)
df1
```

## Data Visualization

### Top 10 Countries with most Terrorist Attacks

```python
df1['Country'].value_counts().head(10)
```

```python
fig, ax = plt.subplots(figsize =(16, 9))
plt.bar(df1['Country'].value_counts()[:10].index,df1['Country'].value_counts()[:10].values,width=0.4)
plt.xlabel("Countries")
plt.ylabel('Count')
plt.title('Top 10 Countries with most Terrorist Attacks')
plt.show()
```

### Top 10 States in India with most Terrorist Attacks

```python
states_data=df1[df1['Country']=='India']['State'].value_counts().head(10)
states_data
```

```python
fig, ax = plt.subplots(figsize =(18, 9))
plt.bar(states_data.index,states_data.values,width=0.4)
plt.xticks(rotation=90)
plt.xlabel('States ')
plt.ylabel('Count')
plt.title("Top 10 States in India with most Terrorist Attacks")
plt.show()
```

### Top 10 Regions with high Terrorism

```python
df1['Region'].value_counts().head(10)
```

```python
fig, ax = plt.subplots(figsize =(18, 9))
plt.bar(df1['Region'].value_counts()[:10].index,df1['Region'].value_counts()[:10].values,width=0.4)
plt.xticks(rotation=90)
plt.xlabel('Regions')
plt.ylabel('Count')
plt.title("Top 10 Regions with high Terrorism")
plt.show()
```

### No. of People Killed over the Years

```python
year_data=df1[['Year','Killed']].groupby('Year').sum()
year_data.head()
```

```python
year_data.plot(kind='bar',figsize =(18, 9))
plt.xlabel('Years')
plt.ylabel('Killed')
plt.title("No. of People Killed over the Years")
plt.show()
```

### Yearwise No. Of People Wounded

```python
year_data1=df1[['Year','Wounded']].groupby('Year').sum()
year_data1.head()
```

```python
year_data1.plot(kind='bar',figsize =(18, 9))
plt.xlabel('Years')
plt.ylabel('Wounded')
plt.title("Yearwise No. Of People Wounded")
plt.show()
```

### Top 15 Cities with most terrorist attacks

```python
df1['City'].value_counts()[1:15]
```

```python
fig,ax=plt.subplots(figsize=(16,8))
plt.bar(df1['City'].value_counts()[1:15].index,df1['City'].value_counts()[1:15].values)
plt.xlabel('Cities')
plt.ylabel('Count')
plt.title("Top 15 Cities with most terrorist attacks")
plt.show()
```

### Most active terrorist groups

```python
df1['Group'].value_counts()[1:10]
```

```python
fig,ax=plt.subplots(figsize=(16,8))
plt.bar(df1['Group'].value_counts()[1:10].index,df1['Group'].value_counts()[1:10].values)
plt.xticks(rotation=90)
plt.xlabel('Groups')
plt.ylabel('Count')
plt.title("Most active terrorist groups")
plt.show()
```
